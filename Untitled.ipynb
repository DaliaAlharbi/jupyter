```python
from ibm_watson import SpeechToTextV1
from ibm_watson.websocket import RecognizeCallback, AudioSource 
from ibm_cloud_sdk_core.authenticators import IAMAuthenticator
```

## 1.Setup STT Service

```python
apikey = 'igwDowHxpKEHG_fnSyNVp4_j4nS24s-CAHdwKaihXKms'
url = 'https://api.eu-gb.speech-to-text.watson.cloud.ibm.com/instances/f3b5bce8-bdd0-47d4-b836-a7dcc583d008'
```

```python
# Setup Service
authenticator = IAMAuthenticator(apikey)
stt = SpeechToTextV1(authenticator=authenticator)
stt.set_service_url(url)
```

## 2. Open Audio Source and Convert

```python
# Perform conversion
with open('my name is dalia.mp3', 'rb') as f:
    res = stt.recognize(audio=f, content_type='audio/mp3', model='en-US_NarrowbandModel', continuous=True).get_result()
```
